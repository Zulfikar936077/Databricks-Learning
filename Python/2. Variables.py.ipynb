## Variables
1. Variables is a container that holds any value like number, text, collection, etc.
2. Variable can store one value at a time means it holds the latest value only
3. Variables are created in memory.

```python
%python
# assigning value to a variable and displaying it
x = 100
print(x)
```

```python
# a. displaying the value of the variable
print(x)
```

```python
# b. displaying the value of the variable
x
```

```python
# Writing multiple-line variable
a = 10
b = 20
c = 30
```

```python
# Adding the values of the variables
print(a)
print(b)
print(c)
print (a + b + c)
```

```python
# Showing the variables in the same line
a1, b1, c1 = 1 , 2, 3
print(a1, b1, c1)
```

```python
# Changing the value of a variable
# For example, the value of x was 100. Now, it is changed to 200.
x = 200
print(x)
```

```python
# Assigning single value to multiple variables
a = b = c = 100
print(a)
print(b)
print(c)
```

```python
# Swaping the values of the variables
a = 10
b = 20
print(a)
print(b)
a,b = b,a
print(a)
print(b)
```
