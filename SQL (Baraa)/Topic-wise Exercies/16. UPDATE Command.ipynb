Attention:
It is always better to use truncate command instead of delete command to remove all the data from the table. Because the delete procedure will take more time and there will be a lot of protocols and logs which will make the process slow. So, it is better to use TRUNCATE.

```sql
TRUNCATE TABLE current_employee_3;
SELECT * FROM current_employee_3;
```
